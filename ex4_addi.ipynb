```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# dataをプロットする関数
def plot_data(x, y, label, color):
    """データをプロットする"""
    plt.plot(x, y, label=label, color=color)
# .datファイルからデータを読み込む
df = pd.read_table('ex4_file/output499.dat', delim_whitespace=True, header=None)

# DをEに置換
df = df.applymap(lambda x: str(x).replace('D', 'E'))

# データ型を変換
df = df.astype(float)

# 2列目(反復回数)をx、1列目(解)をyとして指定,yは解1との差の絶対値を見る
x = df.iloc[:, 1]
dy = df.iloc[:, 0] - 500
y = np.abs(dy)

# 条件を満たすxの値のインデックスを取得
# condition = dy >= 0.0
# indices = np.where(condition)[0]

# グラフの作成
plt.plot(x, 500*0.5**x, color='red')
plt.plot(x, y, marker='o',zorder=1)
plt.xlabel('Times')
plt.ylabel('Error')
plt.title('Error vs Times')
plt.grid(True)
plt.xlabel('times')
plt.ylabel('ans')
plt.title('bisection 15-1015')
# 条件を満たすデータポイントを強調
# plt.scatter(x[indices], y[indices], color='orange',zorder=2)
# 対数軸の設定
plt.yscale('log')
# 両方のグリッド線を引く
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 入力ファイルの設定
input_fname = 'ex6_addi_file/output001.dat'

# 入力ファイルからデータを読み込む,空白を区切りとして読み込む,ヘッダーはなし
df = pd.read_table(input_fname, delim_whitespace=True, header=None)

# DをEに置換
df = df.applymap(lambda x: str(x).replace('D', 'E'))

# データ型を変換
df = df.astype(float)

# 1列目...初期解, 2列目...収束解, 3列目...収束回数
x = df.iloc[:, 0]
y1 = df.iloc[:, 1]
y2 = df.iloc[:, 2]

# グラフの作成
plt.plot(x, y1, marker='o', linestyle='None')
plt.xlabel('x0')
plt.ylabel('x')
plt.title('x vs x0')
# 両方のグリッド線を引く
#plt.grid(True)
plt.show()

# グラフの作成
plt.plot(x, y2, marker='o', linestyle='None')
plt.xlabel('x0')
plt.ylabel('Times')
plt.title('Times vs x0')
# 両方のグリッド線を引く
#plt.grid(True)
plt.show()
```
